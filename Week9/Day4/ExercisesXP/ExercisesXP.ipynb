## Exercise 1 : Basic Scatter Plot With Plotnine

**Objective:** Create a basic scatter plot using Plotnine to visualize the relationship between two variables.

**Tasks:**

- Create a pandas DataFrame with two columns, x and y, each containing 20 random numbers.
- Use Plotnine to create a scatter plot of y versus x.
- Add x and y-axis labels as ‘X-Value’ and ‘Y-Value’, respectively. Also, add a plot title ‘Basic Scatter Plot’. \

**Expected Outcome:** A scatter plot displaying the relationship between the data in x and y, with appropriate labels and title.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  plotnine import ggplot, aes, geom_point, xlab, ylab, ggtitle, geom_bar, facet_wrap, geom_line

df = pd.DataFrame(np.random.randint(1, 100, (20, 2)), columns = ['A', 'B'])
p = (
    ggplot(df)
    + aes(x='A', y='B')
    + geom_point()
    + xlab('X-Value')
    + ylab('Y-Value')
    + ggtitle('Basic Scatter Plot')

)

p.draw()
````

## Exercise 2 : Bar Plot Of Categorical Data With Plotnine

**Objective**: Use Plotnine to create a bar plot showing the frequency of categories in a dataset.

- Generate a pandas DataFrame with a column category, containing 30 random entries from a set of categories (e.g., ‘A’, ‘B’, ‘C’).
- Create a bar plot that shows the count of each category using Plotnine.
- Customize the plot by adding a title ‘Category Frequency’ and x and y-axis labels.

**Expected Outcome:** A bar plot displaying the frequency of each category in the dataset, with clear labels and a descriptive title.

````python
from string import ascii_uppercase as UP
df2 = pd.DataFrame([UP[np.random.randint(0, 10)] for i in range(30)], columns=['Category'])
p2 = (
    ggplot(df2, aes('Category', fill='Category'))
    + geom_bar()
    + xlab('Category')
    + ylab('Frequency')
    + ggtitle('Category Frequency')
)

p2.draw()
````

## Exercise 3 : Line Plot With Multiple Lines With Plotnine

**Objective:** Create a line plot using Plotnine that compares two different data series.

- Create a pandas DataFrame with three columns: x, y1, and y2, where x is a sequence of numbers and y1, y2 are two different mathematical functions of x (e.g., sin(x) and cos(x)).
- Plot y1 and y2 against x in the same plot, using different colors for each line.
- Add a legend to distinguish between the two lines, and label axes as ‘X-Value’, ‘Y-Value’, and add a title ‘Comparison of Two Functions’.

**Expected Outcome:**

A line plot that effectively compares the two functions across the range of x, with a legend, axis labels, and a title.

````python
df3 = pd.DataFrame(np.arange(1, 20), columns = ['x'])
df3['y1'] = np.sin(df3['x'])
df3['y2'] = np.cos(df3['x'])

df3_melted = df3.melt(id_vars='x', value_vars=['y1', 'y2'], var_name='function', value_name='value')
p3 = (
    ggplot(df3_melted, aes(x='x', y='value', color='function'))
    + geom_line()
    + xlab('X-Value')
    + ylab('Y-Value')
    + ggtitle('Comparison of Two Functions')

)
p3.draw()
````

## Exercise 4 : Basic Line Chart With Plotly

Create a basic line chart using Plotly to visualize the trend of sales over 12 months. Assume the following monthly sales data: \
```
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]

```





**Tasks:**

- Import Plotly’s graph_objects as go.
- Create a line chart with months on the x-axis and sales on the y-axis.
- Add a title to the chart: “Monthly Sales Data”.
- Display the plot.

````python
import plotly.express as px
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]
df4 = pd.DataFrame({'Months': months, 'Sales': sales})

fig = px.line(df4, x='Months', y='Sales', title='Monthly Sales Data')
fig.show()
````

## Exercise 5 : Bar Chart With Customization With Plotly

Using Plotly, create a bar chart to represent the number of books sold in different genres. Use the following data:



```
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
```
**Tasks:**

- Import Plotly’s graph_objects as go.
- Create a bar chart with genres on the x-axis and books_sold on the y-axis.
- Customize the bar colors.
- Add a title to the chart: “Books Sold by Genre”.
- Display the plot.
