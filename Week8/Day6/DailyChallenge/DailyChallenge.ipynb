```python
import pandas as pd
data = pd.read_csv('2019.csv')
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Social support'], data['Score'])
plt.xlabel('Social support')
plt.ylabel('Score')
```

```python
fig, ax1 = plt.subplots(figsize=(22,12))

ax1.bar(data['Country or region'], data['GDP per capita'], color='SkyBlue', label='GDP per Capita')
ax1.set_xlabel('Region')
ax1.set_ylabel('GDP per Capita')
ax1.set_title('GDP per Capita and Healthy Life Expectancy by Region (2019)')
ax1.tick_params(axis='x', rotation=90)

ax2 = ax1.twinx()
ax2.plot(data['Country or region'], data['Healthy life expectancy'], color='IndianRed', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy')

fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
```
