```python
import pandas as pd
df = pd.read_csv('IRIS.csv')
df['sepal_length'][21] = None
df['sepal_width'][38] = None
df['petal_length'][13] = None
df['sepal_length'][89] = None
pd.notna(df)
```

```python
values = {'sepal_length': df['sepal_length'].median(),'sepal_width':	df['sepal_width'].median(),'petal_length':	df['petal_length'].mean()}
df.fillna(values)
```

```python
df.rename(columns={'sepal_length': 'Sepal length, cm'	,'sepal_width':	'Sepal width, cm','petal_length':	'Petal length, cm', 'petal_width': 'Petal width, cm'})
```

```python
df.to_excel('cleaned_iris.xlsx', index=False)
```

```python
df_excel = pd.read_excel('cleaned_iris.xlsx')
```

```python
df_excel.to_json('iris_subset.json', orient='records')
```
