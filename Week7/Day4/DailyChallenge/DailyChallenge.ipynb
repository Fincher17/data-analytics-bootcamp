```python
import matplotlib.pyplot as plt
from PIL import Image as img, ImageOps
import numpy as np

image = np.asarray(Image.open('/content/flowers/flowers/00_002.png'))
aug = np.asarray(Image.open('/content/flowers/flowers/00_002.png'))


def show_images(original, augmented, title):
  plt.figure(figsize=(12, 6))
  plt.subplot(1, 2, 1)
  plt.imshow(original)
  plt.title('Original Image')
  plt.axis('off')

  plt.subplot(1, 2, 2)
  plt.imshow(augmented)
  plt.title(title)
  plt.axis('off')

  plt.show()

rot90 = np.rot90(aug, 3)
show_images(image, rot90, 'Rotate by 90 degrees')
flip_v = np.flip(aug, axis=0)
show_images(image, flip_v, 'vertically fliped image')
flip_h = np.flip(aug, axis=1)
show_images(image, flip_h, 'horizontally fliped image')
res = img.open(r'/content/flowers/flowers/00_002.png')
print(res)
res = res.resize((154, 154))
show_images(image, res, 'scale by 1.2x')
```
